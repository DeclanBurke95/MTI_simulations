```python
from systems import make_mti_lead
from funcs import bands
from funcs import get_default_lead_params
import kwant
import numpy as np
import matplotlib.pyplot as plt
```

```python
lead=make_mti_lead(400, ph_symmetry=False)
params=get_default_params(finite=False)
params.update(mu=0.02)
```

```python
bands=kwant.physics.Bands(lead, params=params)
momenta=np.linspace(-0.35,0.35,301)
energies_normal=[bands(k) for k in momenta]

x1=np.array(energies_normal)
en1=(x1 *1000)
plt.xlim(-0.35,0.35)
plt.ylim(-100,100)

plt.plot(momenta, en1, color='k')
plt.ylabel('Energy (meV)', fontsize=12)
plt.xlabel('k', fontsize=12)
plt.show()
```
