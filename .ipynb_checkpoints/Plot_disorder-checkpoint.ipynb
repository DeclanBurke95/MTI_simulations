```python
import matplotlib.pyplot as plt
from systems import get_disorder2
from systems import syst_create_mti
import numpy as np
```

```python
# Plot a 2D colormap of the disorder profile
def plot_2d_disorder(L, W, S_mag, n_dis, corrL, L_normal):
    fsyst=syst_create_mti(L, W)
    disorder=get_disorder2(n_dis, S_mag, corrL, L, W, L_normal)
    
    x=np.linspace(0,int((L/10)),int((L/10)+1))
    y=np.linspace(0,int((W/10)),int((W/10)+1))
    X,Y=np.meshgrid(x,y)
    plt.pcolormesh(X, Y, disorder)
```

```python
# Plot a 1D plot of disorder, averaged over y-coordinates 
def plot_1d_disorder(L, W, S_mag, n_dis, corrL, L_normal):
    fsyst=syst_create_mti(L, W)
    disorder=get_disorder2(n_dis, S_mag, corrL, L, W, L_normal)
    
    x=np.linspace(0,int((L/10)),int((L/10)+1))
    plt.plot(x,disorder.mean(axis=0))
```
