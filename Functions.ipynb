```python
def create_MTI():
    W=W
    L=L
    a = 10
    syst = kwant.Builder()
    
    hamiltonian_MTI = (
    "hbar * vF * k_y * kron(sigma_x, sigma_z) - "
    "hbar * vF * k_x * kron(sigma_y, sigma_z) + "
    "{m} * kron(sigma_0, sigma_x) + " # top-bottom surface coupling
    "Msysz * kron(sigma_z, sigma_0)")
    
    hamiltonian_MTI = hamiltonian_MTI.format(m=m)
    ham_MTI_sys_discretized = kwant.continuum.discretize(hamiltonian_MTI, grid=a)
    ham_MTI_leads_discretized = kwant.continuum.discretize(hamiltonian_MTI, grid=a)

    syst.fill(ham_MTI_sys_discretized, *shape(L,W))

    lead_left = kwant.Builder(kwant.TranslationalSymmetry((-a, 0)))
    lead_right = kwant.Builder(kwant.TranslationalSymmetry((a, 0)))  
    lead_left.fill(ham_MTI_lead_discretized, *shape_pointcontact(W))
    lead_right.fill(ham_MTI_lead_discretized, *shape(L, W))

    print(syst)
    syst.attach_lead(lead_left)
    syst.attach_lead(lead_right)

    fsyst=syst.finalized()

a = 10
syst = kwant.Builder()

NEE=[]
REE=[]
REH=[]

Nvals=np.linspace(0,36,37)
Lvals=np.linspace(2000, 20000, 37)

parser = argparse.ArgumentParser(description='E(L) plot')
parser.add_argument('N', type=float, help='Nanoribbon Length')
args = parser.parse_args()
N=int(args.N)
nw=int(Nvals[N-1])

W=400
L=int(Lvals[nw])

a = 10
syst = kwant.Builder()

term_SC_imp = (
    " + S_imp(site, Smag_imp, disorder_imp) * kron(sigma_z, sigma_0, sigma_0)"   
) 

hamiltonian_MTI = (
    "hbar * vF * k_y * kron(sigma_x, sigma_z) - "
    "hbar * vF * k_x * kron(sigma_y, sigma_z) + "
    "{m} * kron(sigma_0, sigma_x) + " # top-bottom surface coupling
    "Msysz * kron(sigma_z, sigma_0)"
)

hamiltonian_tip =  (
    "(-mu_m + V) * kron(sigma_z, sigma_0, sigma_0) + "
    "C_m * (k_x**2 + k_y**2) * kron(sigma_z, sigma_0, sigma_0)"
)

hamiltonian_SMTI = (
    "-mu * kron(sigma_z, sigma_0, sigma_0) +"
    "hbar * vF * k_y * kron(sigma_0, sigma_x, sigma_z) - "
    "hbar * vF * k_x * kron(sigma_z, sigma_y, sigma_z) + "
    "{m} * kron(sigma_z, sigma_0, sigma_x) + " # top-bottom surface coupling
    "Msysz * kron(sigma_z, sigma_z, sigma_0) + " # magnetization term
    " - re({Delta}) * (kron(sigma_y, sigma_y,sigma_0) + kron(sigma_y, sigma_y,sigma_z))/2 - "
    "im({Delta}) * (kron(sigma_x, sigma_y,sigma_0) + kron(sigma_x, sigma_y,sigma_z))/2"

    )

hamiltonian_SMTI_lead = (
    "-mu * kron(sigma_z, sigma_0, sigma_0) +"
    "hbar * vF * k_y * kron(sigma_0, sigma_x, sigma_z) - "
    "hbar * vF * k_x * kron(sigma_z, sigma_y, sigma_z) + "
    "{m} * kron(sigma_z, sigma_0, sigma_x) + " # top-bottom surface coupling
    "Msysz * kron(sigma_z, sigma_z, sigma_0)")

m = "(m0 - m1 * (k_x**2 + k_y**2))"

hamiltonian_SMTI = hamiltonian_SMTI.format(m=m, Delta="Delta(x)")
hamiltonian_SMTI_lead = hamiltonian_SMTI_lead.format(m=m)
hamiltonian_MTI = hamiltonian_MTI.format(m=m)

ham_MTI_sys_discretized = kwant.continuum.discretize(hamiltonian_MTI, grid=a)
ham_MTI_leads_discretized = kwant.continuum.discretize(hamiltonian_MTI, grid=a)
ham_SMTI_sys_discretized = kwant.continuum.discretize(hamiltonian_SMTI + term_SC_imp, grid=a)
ham_SMTI_lead_discretized = kwant.continuum.discretize(hamiltonian_SMTI_lead, grid=a)

ham_tip_discretized = kwant.continuum.discretize(hamiltonian_tip, grid=a)

def conductance_old(fsyst, energy, params, lead_in, lead_out):
      m_eff=0.1
      params.update(dict(C_m=3.81/m_eff))
      smatrix = kwant.smatrix(fsyst, energy, params=params)
      N_e = smatrix.submatrix((lead_in, 0), (lead_in, 0)).shape[0]
      r_ee = smatrix.transmission((lead_out, 0), (lead_in, 0))
      r_he = smatrix.transmission((lead_out, 1), (lead_in, 0))

      return N_e, r_ee, r_he

tau_z = np.kron(np.kron(np.array([[1, 0], [0, -1]]), np.eye(2)), np.eye(2))

def shape(L, W, v=np.array([0,0])):
    L_start, W_start = 0, 0
    L_stop, W_stop = L, W

    def _shape(site):
        (x, y) = site.pos - v
        return (0 <= x <= L and 0 <= y <= W)

    return _shape, v + np.array([L_stop, W_stop])

def shape_pointcontact(W):
    def _shape(site):
        (x, y) = site.pos
        return y==W//2

    return _shape, (0,W//2)

syst.fill(ham_SMTI_sys_discretized, *shape(L,W))

lead_left = kwant.Builder(kwant.TranslationalSymmetry((-a, 0)), conservation_law=-tau_z)
# lead_left = kwant.Builder(kwant.TranslationalSymmetry((-a, 0)))
lead_right = kwant.Builder(kwant.TranslationalSymmetry((a, 0)))  
lead_left.fill(ham_tip_discretized, *shape_pointcontact(W))
lead_right.fill(ham_SMTI_lead_discretized, *shape(L, W))

print(syst)
syst.attach_lead(lead_left)
syst.attach_lead(lead_right)

fsyst=syst.finalized()

n_corrL = 2
delta_corrL = 20
# L_list=[0,20]
corrL_list = [delta_corrL*ind for ind in range(n_corrL)]

n_dis = 200
L_normal=1000

disorder_list = [[] for corrL in corrL_list]
disorder = []

for ind_corrL, correlation_length in enumerate(corrL_list):
    for ind in range(n_dis):
        if (correlation_length >= 1):
            x, y = np.mgrid[-3*correlation_length:3*correlation_length+1:1, -3*correlation_length:3*correlation_length+1:1]
            pos = np.stack((x, y), axis=2)
            rv = multivariate_normal([0, 0], [[correlation_length**2, 0], [0, correlation_length**2]])
            filter_kernel = rv.pdf(pos)

            disorder = np.array([[kwant.digest.gauss('('+str(ind_x) + ',' + str(ind_y) +')', salt=str(ind))
                                   for ind_x in range(L//a+1)]
                                 for ind_y in range(W//a+1)])
            disorder = scipy.signal.fftconvolve(disorder, filter_kernel, mode='same')
            disorder = disorder/np.std(disorder.flatten())
        else:
            disorder = np.array([[kwant.digest.gauss('('+str(ind_x) + ',' + str(ind_y) +')', salt=str(ind))
                                   for ind_x in range(L//a+1)]
                                for ind_y in range(W//a+1)])

        for ind_x in range(L_normal//a+1):
            for ind_y in range(W//a+1):
                disorder[ind_y, ind_x] = 0                    
        disorder_list[ind_corrL].append(disorder)

def get_S_imp(salt):

    def S_imp(site, Smag_imp, disorder_imp):
        ind_x = int(site.pos[0]/a)
        ind_y = int(site.pos[1]/a)
        if (ind_x >= disorder_imp.shape[1]) or (ind_y >= disorder_imp.shape[0]):
            #print('Out of bounds. No disorder applied to site.')
            return 0
        else:
            return Smag_imp * disorder_imp[ind_y, ind_x]
    return S_imp

def get_disorder_imp(corrL, salt):
    def condition_corrL(x): return x == corrL
    def condition_corrL_default(x): return x == 0
    try:
        ind_corrL = [idx for idx, element in enumerate(corrL_list) if condition_corrL(element)][0]
    except:
        print('Correlation length not in range. Defaulted to 0.')
        ind_corrL = [idx for idx, element in enumerate(corrL_list) if condition_corrL_default(element)][0]
    try:
        ind_salt = int(salt)
    except ValueError:
        print('Salt string not an integer. Defaulted to 0.')
        ind_salt = 0
    disorder_imp = disorder_list[ind_corrL][ind_salt]
    return disorder_imp


params = dict(hbar=1,
                    mu=0.014823834205238105,
                    V=0,
                    mu_m=1.6,
                    vF=3,#for Bi2Se3-based 3DTI, 10^15 canceled out with hbar
                    m0=-0.005,
                    m1=15,
                    Msysz=0.015,
                    Msysx=.0,
                    Msysy=.0,
                    Mz=0.1,
                    Mx=.0,
                    My=.0,
                    a=a,
                    phi_0=1.0,
                    exp=np.exp,
                    Smag_imp=0.003,
                    Smag_mimp=.0,
                    Stheta_mimp=np.pi/2,
                    Sphi_mimp=.0,
                    disorder_imp=np.zeros([1, 1]),
                    correlation_length=35,
                    re=lambda x: x.real,
                    im=lambda x: x.imag,
                    Delta=lambda x: 0.005 if x >= 1000 else 0,
                    S_imp=get_S_imp("")
                    )

for n in range(200):
    params.update(correlation_length=20)
    params.update(disorder_imp=get_disorder_imp(params['correlation_length'], str(n)))
    NEE.append(conductance_old(fsyst, 0.0, params, 0, 0)[0])
    REE.append(conductance_old(fsyst, 0.0, params, 0, 0)[1])
    REH.append(conductance_old(fsyst, 0.0, params, 0, 0)[2])

nee=np.array(NEE)
ree=np.array(REE)
reh=np.array(REH)

TC=nee-ree+reh

np.savetxt(f"TCvsL_cor20{N}.csv", TC, delimiter=',')
```
